# Introduction

This notebook simulates networks of excitatory and inhibitory integrate-and-fire neurons. It replicates figures from the following paper:

Brunel, Nicolas. "Dynamics of sparsely connected networks of excitatory and inhibitory spiking neurons." _Journal of computational neuroscience_ 8 (2000): 183-208.

#### Figure 8A

```python
from brian2 import *
from brian2tools import *

# Network parameters
Vr = 10*mV
theta = 20*mV
tau = 20*ms
delta = 1.5*ms
taurefr = 2*ms
duration = .5*second

sparseness = 0.1
Je = .1*mV
Ji = .3*mV

muext = 40*mV
sigmaext = 2.1*mV

# Neuron model
eqs = "dV/dt=(-V+muext+sigmaext*sqrt(tau)*xi)/tau : volt (unless refractory)"
N = NeuronGroup(12500, eqs, threshold='V>theta', reset='V=Vr', refractory=taurefr, method='euler')

# Populations
N.V = Vr
Ne = N[:10000] # Ne = N.subgroup(10000)
Ni = N[10000:] # Ni = N.subgroup(2500)
X = Ne[:50] # X = Ne.subgroup(50)

#Synapses
Ce1 = Synapses(Ne, Ne, on_pre='V += Je', delay=delta)
Ce1.connect(p=sparseness)

Ce2 = Synapses(Ne, Ni, on_pre='V += Je', delay=delta)
Ce2.connect(p=sparseness)

Ci1 = Synapses(Ni, Ni, on_pre='V += -Ji', delay=delta)
Ci1.connect(p=sparseness)

Ci2 = Synapses(Ni, Ne, on_pre='V += -Ji', delay=delta)
Ci2.connect(p=sparseness)

# Monitors
M = PopulationRateMonitor(N)
S = SpikeMonitor(N)

# Run
run(duration)

# Plot
subplot(211)
plt.plot(M.t/ms,M.rate, color='k')
plt.xlim(400,500)
plt.xlabel('time (ms)')
plt.ylabel('firing rate (Hz)')

subplot(212)
raster_plot = plot_raster(S.i, S.t, time_unit=ms, color='k')
raster_plot.set_xlim(400,500)
raster_plot.set_ylim(0,50)
show()
```
